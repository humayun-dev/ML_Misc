```python
# to demonstrate the train_test_split concept
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("/content/drive/MyDrive/datasets/add.csv")
data
# in the output below, x and y are independent variables while the sum is the dependent variable
# total data equals to the 22 rows
# lets train the data 80% and 20% for the test
# from 0 to 15 equals x_train(x,y)
# from 0 to 15 equals y_train(sum)
# from 16 to 22 equals x_test(x,y)
# from 16 to 22 equals y_test(sum)
```

```python
X = data[['x','y']] # assign the x-train or independent values
y = data['sum']     # assign the y-train or the dependent values

X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
X_train
```

```python
X_test
```

```python
# train the model, as the values are the continous so LinearRegression will be used
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

# now compare the results using X_test and y_test
y_pred = model.predict(X_test) # the X_test both values will be added and result is stored in the y_predict
y_pred
```

```python
# lets check out the y_test values for the comparison
y_test  # the following output predicts the correct model training
```

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```
