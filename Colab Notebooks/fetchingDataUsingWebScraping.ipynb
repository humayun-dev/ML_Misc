```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get("https://www.uetmardan.edu.pk",headers = headers).text  # all information of the webpage to get
```

```python
# now using BeatifulSoup parse the XML into the HTML
soup = BeautifulSoup(webpage,'lxml')
print(soup.prettify())  # prettify prints the information in the HTML format
```

```python
# now to find out how many headers in the website
soup.find_all('h2')
```

```python
# find out the h2 header in the text format
soup.find_all('h2')[0].text
```

```python
# to print all the paragraph tags in the website
for i in soup.find_all('p'):
  print(i.text.strip())   # strip function is used to ignore all the spaces etc. in the text
```
