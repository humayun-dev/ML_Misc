```python
# to demonstrate the linear regression
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/datasets/employee.csv")
df
```

```python
xpoints = df["Experience"]  # independent variable
ypoints = df["Salary"]      # dependent variable
plt.plot(xpoints,ypoints)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# create a linear regression model
model = linear_model.LinearRegression()

# As salary is the target column and needs to be drop for the model to train
train_df = df.drop("Salary",axis = 1)
train_df
```

```python
model.fit(train_df,df['Salary'])
```

```python
model.predict([[12]]) # what if the experience is 12 years, the prediction is the salary 1735121
```
